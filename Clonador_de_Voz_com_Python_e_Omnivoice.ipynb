## Importação das Biblkotecas

```python
import torch
import soundfile as sf
from pydub import AudioSegment
from omnivoice import OmniVoice
import os
from google.colab import files
```

## Entrada do texto e modelo de voz.wav para clonar

```python
uploaded = files.upload()
voice_path = list(uploaded.keys())[0]
texto = input("Digite o texto: ")
```

## Carregar Modelo

```python
print("🔹 Carregando modelo...")
model = OmniVoice.from_pretrained(
    "k2-fsa/OmniVoice",
    device_map="cuda:0" if torch.cuda.is_available() else "cpu",
    dtype=torch.float16 if torch.cuda.is_available() else torch.float32
)
```

## Gerar Áudio

```python
print("🎙️ Gerando áudio...")
audio = model.generate(
    text=texto,
    ref_audio=voice_path
)
```

## Salvar Áudio

```python
os.makedirs("saida", exist_ok=True)

wav_path = "saida/output.wav"
mp3_path = "saida/output.mp3"

sf.write(wav_path, audio[0], 24000)

AudioSegment.from_wav(wav_path).export(mp3_path, format="mp3")

print("✅ Áudio gerado em:", mp3_path)
```
